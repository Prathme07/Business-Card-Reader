```python
dataList = list(map(lambda x: x.split("\t"), data.split("\n")))
```

```python
dataList
```

```python
df = pd.DataFrame(dataList[1:], columns=dataList[0])

df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True) # it will remove the missing values in rows

col_int = ["level", "page_num", "block_num", "par_num", "line_num", "word_num", "left", "top", "width", "height", "conf"]
df[col_int] = df[col_int].astype(int)
```

```python
df.dtypes
```

```python
df.info()

df.head()
```

```python
image = img.copy()

level = "word"
for l, x, y, w, h, c, txt in df[["level", "left", "top", "width", "height", "conf", "text"]].values:
    #print(l, x, y, w, h, c)
    if level == "page":
        if l == 1:
            cv2.rectangle(image, (x, y),(x+w, y+h), (0,0,0),3) # black color
        else:
            continue
            
    elif level == "block":
        if l == 2:
            cv2.rectangle(image, (x, y),(x+w, y+h), (255,0,0),2) # blue color
        else:
            continue
            
            
    elif level == "para": # paragraph
        if l == 3:
            cv2.rectangle(image, (x, y),(x+w, y+h), (0,255,0),2) # green color
        else:
            continue
            
    elif level == "line": # lines
        if l == 4:
            cv2.rectangle(image, (x, y),(x+w, y+h), (0,0,255),2) # red color
        else:
            continue
            
    elif level == "word": # paragraph
        if l == 5:
            cv2.rectangle(image, (x, y),(x+w, y+h), (0,255,0),2) # green color
            cv2.putText(image, txt, (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0), 2)
        else:
            continue
            
plt.imshow(image)
plt.axis("off")
```
